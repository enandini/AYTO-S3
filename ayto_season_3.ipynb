# Preliminarys

```
import itertools
import numpy as np
import math
import pandas as pd

list1=["Hannah", "Rashida", "Amanda", "Melanie", "Chelsey", "Kiki", "Stacy", "Britni", "Cheyene", "Kayla"]
list2=["Chuck", "Tyler", "Mike", "Devin", "Connor", "Austin", "Alec", "Hunter", "Nelson", "Zak"]

perms = [list(zip(x,list2)) for x in itertools.permutations(list1,len(list2))]
```

```
truth_booths = []
matchups = []

print("Starting matches: " + str(math.factorial(10)))
```

```
def check_truth_booths(perm, tbs):
  for tb in tbs:
    if tb[1] == False and tb[0] in perm:
      return False
    elif tb[1] == True and tb[0] not in perm:
      return False    
  return True
```

```
def check_matchups(perm, ms):
  for m in ms:
    limit = m[-1]
    matches = 0
    for pair in perm:
      if pair in m:
        matches += 1
    if matches != limit:
      return False
  return True
```

## Week 1 Data

```
truth_booths.append([("Kiki","Hunter"), False])
matchups.append([("Hannah","Chuck"),
                 ("Rashida", "Tyler"),
                 ("Amanda", "Mike"),
                 ("Melanie", "Devin"),
                 ("Chelsey", "Connor"),
                 ("Kiki", "Austin"),
                 ("Stacy", "Alec"),
                 ("Britni", "Hunter"),
                 ("Cheyene", "Nelson"),
                 ("Kayla", "Zak"),2])
```

```
match_dict = {}

for i in list1:
    match_dict[i] = {}
    for j in list2:
        match_dict[i][j] = 0

valid_perms = 0
for i in perms:
  if check_truth_booths(i, truth_booths) == False:
    pass
  elif check_matchups(i, matchups) == False:
    pass
  else:
    for pair in i:
      match_dict[pair[0]][pair[1]] += 1
    valid_perms += 1

print(str(valid_perms) + " remaining permutations.")
np.round((pd.DataFrame(match_dict)/valid_perms)*100,2)
```
